```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import sys
```

```python
states = df['STATE'].unique()

def lda_analysis_by_state(df, state, n_topics=3):
    # Filtrar los datos para el estado
    df_state = df[df['STATE'] == state]
    
    # Vectorizaci√≥n con TfidfVectorizer
    vectorizer = TfidfVectorizer(stop_words='english', max_df=0.95, min_df=2)
    X = vectorizer.fit_transform(df_state['tweet'])
    
    # Aplicar LDA
    lda = LatentDirichletAllocation(n_components=n_topics, random_state=42, max_iter=10, verbose=1)
    lda.fit(X)
    
    # Mostrar los t√©rminos m√°s importantes para cada tema
    for index, topic in enumerate(lda.components_):
        print(f'Tema {index}:')
        print([vectorizer.get_feature_names_out()[i] for i in topic.argsort()[-5:]])  # Las 5 palabras m√°s relevantes
        print('\n')

# Ejecutar LDA para un estado 
lda_analysis_by_state(df, 'AZ', n_topics=3)
```
